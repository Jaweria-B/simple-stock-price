```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# MSFT
# AAPL
# GOOGL
tickerSymbol = 'GOOGL'
tickerData = yf.Ticker(tickerSymbol)
#tickerData
```

```python
tickerDF = tickerData.history(period='1d', start='2013-5-31', end='2023-5-31')
```

```python
plt.plot(tickerDF.Close, c='blue')
plt.title(f'{tickerSymbol} Stocks')
plt.show()
```

```python
plt.plot(tickerDF.Volume, c='blue')
plt.title(f'{tickerSymbol} Stocks')
plt.show()
```
